```python
import os
import random
import shutil

# === Input Dataset Paths ===
structural_pos = 'D:/MajorProject/train/Positive'
structural_neg = 'D:/MajorProject/train/Negative'
land_pos = 'D:/MajorProject/data_land/crack'
land_neg = 'D:/MajorProject/data_land/nocrack'

# === Output Combined Dataset ===
combined_dir = 'D:/MajorProject/combined_data'
pos_dir = os.path.join(combined_dir, 'positive')
neg_dir = os.path.join(combined_dir, 'negative')

# Create directories
os.makedirs(pos_dir, exist_ok=True)
os.makedirs(neg_dir, exist_ok=True)

# === Helper Function ===
def get_images(folder):
    return [os.path.join(folder, f) for f in os.listdir(folder)
            if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

# === Load Paths ===
struct_pos_list = get_images(structural_pos)
struct_neg_list = get_images(structural_neg)
land_pos_list = get_images(land_pos)
land_neg_list = get_images(land_neg)

# === Balance Counts ===
min_pos = min(len(struct_pos_list), len(land_pos_list))
min_neg = min(len(struct_neg_list), len(land_neg_list))

print(f"⚖️ Using {min_pos} positive and {min_neg} negative samples from each domain")

random.seed(42)
struct_pos_list = random.sample(struct_pos_list, min_pos)
land_pos_list = random.sample(land_pos_list, min_pos)
struct_neg_list = random.sample(struct_neg_list, min_neg)
land_neg_list = random.sample(land_neg_list, min_neg)

# === Combine ===
all_pos = struct_pos_list + land_pos_list
all_neg = struct_neg_list + land_neg_list

# === Copy Files ===
def copy_files(file_list, target_folder):
    for src in file_list:
        dst = os.path.join(target_folder, os.path.basename(src))
        shutil.copy(src, dst)

copy_files(all_pos, pos_dir)
copy_files(all_neg, neg_dir)

print("✅ Combined dataset created successfully!")
print(f"Total Positives: {len(all_pos)}")
print(f"Total Negatives: {len(all_neg)}")
```
